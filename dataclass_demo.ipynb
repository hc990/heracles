```python
from dataclasses import dataclass

a = []
for i in range(5):
    a.append(i)  # Simply append without reassignment
b = a[1:3]
print(a)
print(b)
    
@dataclass(order=True)
class Person:
    name: str
    age: int
   
person = Person("Lee",30)
person2 = Person("Dee",30)

print(person < person2)

print(person)
```
