```python
import asyncio
from time import sleep, perf_counter

##定义携程函数
##包装携程任务
##建立事件循环

# async def fetch_url(url):
#     print('fetch url')
#     await asyncio.sleep(1) ##await 包装成任务
#     print('finish fetch url')
#     return 'url good'
    
# async def read_book(bookname):
#     print('read_book')
#     await asyncio.sleep(1) ##await 包装成任务
#     print('finish read_book')
#     return 'book good'

# async def main():
#     url ="-----"
#     bookname='books'
#     task1 = asyncio.create_task(fetch_url(url))
#     task2 = asyncio.create_task(read_book(bookname))
#     await asyncio.gather(task1, task2, return_exceptions=True)
#     fetch_result = await task1
#     read_result = await task2
    
    
# if __name__ == '__main__':
#     start_time = perf_counter()
#     asyncio.run(main())
#     end_time = perf_counter()
#     print(f'Time taken: {end_time-start_time:.2f} seconds')
    
    
# async def main():
#     url = "https://api.pmtcgo.com/commodities"
#     bookname = 'Pokémon Guide'
#     task1 = asyncio.create_task(fetch_url(url))
#     task2 = asyncio.create_task(read_book(bookname))
#     fetch_result = await task1
#     read_result = await task2
#     results = await asyncio.as_completed(task1, task2, return_exceptions=True)
#     fetch_result, read_result = results
#     print(f'Fetch result: {fetch_result}')
#     print(f'Read result: {read_result}')

# # 在 Jupyter Notebook 或 IPython 中运行
# start_time = perf_counter()
# await main()  # 直接 await，不用 asyncio.run()
# end_time = perf_counter()
# print(f'Time taken: {end_time - start_time:.2f} seconds')
    
    
import asyncio
from time import perf_counter
import aiohttp

async def fetch_url(url: str) -> str:
    print('Fetching URL:', url)
    await asyncio.sleep(1)  # 模拟延迟
    print('Finished fetching URL:', url)
    return f'URL {url} response: 200'
   

async def read_book(bookname: str) -> str:
    print('Reading book:', bookname)
    await asyncio.sleep(1)
    print('Finished reading book:', bookname)
    return f'Book {bookname} is good'

async def main():
    url = "https://api.pmtcgo.com/commodities"
    bookname = 'Pokémon Guide'

    task1 = asyncio.create_task(fetch_url(url))
    task2 = asyncio.create_task(read_book(bookname))

   
    for coro in asyncio.as_completed([task1, task2]):
        result = await coro
        print("Task finished:", result)
#     print(f'Fetch result: {fetch_result}')
#     print(f'Read result: {read_result}')

if __name__ == '__main__':
    start_time = perf_counter()
    await main()
    end_time = perf_counter()
    print(f'Time taken: {end_time - start_time:.2f}  seconds')
```
