### pytorch 函数

#### torch.cat

功能:将一组张量（tensors）沿指定维度（dim）拼接成一个新张量。张量除拼接维度外，其余维度形状必须相同。
参数:
tensors：要拼接的张量序列（如列表或元组），通常为 torch.Tensor 类型，数量至少为 2。
dim：拼接的维度（整数，默认 0）。例如，dim=0 按行拼接（增加行数），dim=1 按列拼接（增加列数）。
out（可选）：输出张量的存储位置（默认 None，返回新张量）。
注意：张量的数据类型（dtype）和设备（device，如 CPU/GPU）必须一致，否则需先转换（如用 .to(device)）。 

#### torch.nn.BatchNorm2d

nn.BatchNorm2d 是一种批量归一化（Batch Normalization，简称 BN）层，专门用于处理二维输入数据（如卷积神经网络中的特征图）。它通过对每个小批量（mini-batch）的数据进行归一化和缩放来加速训练、稳定梯度并提高模型性能。

nn.BatchNorm2d 对输入特征图在每个通道上进行归一化，具体步骤为：

标准化：将每个通道的数据标准化为均值为 0、方差为 1。
缩放和平移：通过可学习的参数（缩放因子 $\gamma$ 和偏置 $\beta$）对标准化后的数据进行线性变换，恢复模型的表达能力。

主要作用:

加速训练：通过标准化输入，减少内部协变量偏移（Internal Covariate Shift），使梯度更稳定，允许使用更高的学习率。
正则化效果：小批量统计的随机性引入噪声，类似 Dropout，具有轻微正则化作用，减少过拟合。
稳定梯度：归一化使激活值分布更稳定，缓解梯度消失/爆炸问题，适合深层网络。
减少对初始化的依赖：BatchNorm 使模型对权重初始化的敏感性降低，便于训练。

是一个容器类，用于按顺序组合多个神经网络层（或模块），以简化模型定义和前向传播的实现。它允许开发者将一系列层按序“堆叠”，数据会依次通过每个层进行处理。以下是对 nn.Sequential 作用的简洁而全面的解释，结合使用场景和代码示例。


#### torch.nn.Conv2d 

是一个用于二维卷积操作的模块,通过在输入张量上滑动卷积核（kernel），执行卷积操作，生成输出特征图（feature map）。

1. 基本作用
nn.Conv2d 通过在输入张量上滑动卷积核（kernel），执行卷积操作，生成输出特征图（feature map）。它能够：

提取局部特征：卷积核捕捉输入的局部区域特征（如图像中的边缘或角点）。
降维或升维：通过调整输出通道数，改变特征图的维度。
共享参数：卷积核在整个输入上共享权重，减少参数量，提高计算效率。

数学上，卷积操作对输入张量 $ x $ 和卷积核 $ w $ 进行如下计算（忽略偏置）：
$$y[i, j, c_{out}] = \sum_{c_{in}} \sum_{m, n} x[i \cdot s + m, j \cdot s + n, c_{in}] \cdot w[m, n, c_{in}, c_{out}] + b[c_{out}]$$

$ x $: 输入张量，形状为 $[batch_size, C_{in}, H_{in}, W_{in}]$。
$ w $: 卷积核，形状为 $[C_{out}, C_{in}, kernel_size, kernel_size]$。
$ b $: 偏置，形状为 $[C_{out}]$。
$ y $: 输出张量，形状为 $[batch_size, C_{out}, H_{out}, W_{out}]$。
$ s $: 步幅（stride）。


2. 主要参数
创建 nn.Conv2d 实例时，常用参数包括：

in_channels: 输入张量的通道数（如 RGB 图像为 3）。 
    
out_channels: 输出特征图的通道数（即卷积核数量，输出的特征数量）。决定了卷积层提取的特征种类数量，每个通道对应一种特征表示。

kernel_size: 卷积核大小（如 3 表示 $3 \times 3$）。 池化核的大小（如 2 或 $(2, 2)$ 表示 $2 \times 2$ 窗口）。决定是几 * 几的卷积

stride: 卷积核滑动的步幅（默认 1）。

padding: 输入边缘填充的像素数（默认 0，常设为 (kernel_size-1)/2 以保持尺寸）。

bias: 是否添加偏置（默认 True）。


3. 输出维度公式
输出特征图的维度由以下公式决定：
$$H_{out} = \left\lfloor \frac{H_{in} + 2 \cdot padding - kernel_size}{stride} \right\rfloor + 1$$
$$W_{out} = \left\lfloor \frac{W_{in} + 2 \cdot padding - kernel_size}{stride} \right\rfloor + 1$$

输出形状：$[batch_size, out_channels, H_{out}, W_{out}]$。
如果 padding = (kernel_size-1)/2（对于奇数 kernel_size），且 stride=1，则 $H_{out} = H_{in}$，$W_{out} = W_{in}$，保持空间尺寸不变。


#### nn.MaxPool2d

torch.nn.MaxPool2d 是一种二维最大池化（Max Pooling）层，广泛用于卷积神经网络（CNN）中。它的主要作用是对输入特征图进行下采样（downsampling），通过选择局部区域中的最大值来减少空间维度（高度和宽度），同时保留重要特征。

1. 基本作用
nn.MaxPool2d 在输入张量的每个通道上，应用一个滑动窗口（池化核），从窗口覆盖的区域中选择最大值作为输出，从而生成尺寸较小的特征图。其核心功能包括：

降维：减少特征图的空间尺寸（如高度和宽度），降低计算量和参数量。
保留显著特征：通过选择最大值，保留局部区域中最强的特征（如边缘、角点），增强模型对平移、旋转等变化的鲁棒性。
防止过拟合：降维减少模型复杂度，起到一定的正则化作用。

2. 数学原理
对于输入张量 $[batch_size, C, H_{in}, W_{in}]$（其中 $C$ 是通道数，$H_{in}$ 和 $W_{in}$ 是高度和宽度），nn.MaxPool2d 在每个通道独立操作：

在每个通道上，池化核（大小为 kernel_size）以步幅 stride 滑动，选取窗口内最大值。
输出张量形状为 $[batch_size, C, H_{out}, W_{out}]$，其中：
$$H_{out} = \left\lfloor \frac{H_{in} + 2 \cdot padding - kernel_size}{stride} \right\rfloor + 1$$
$$W_{out} = \left\lfloor \frac{W_{in} + 2 \cdot padding - kernel_size}{stride} \right\rfloor + 1$$

操作示例：假设 $kernel_size=2, stride=2, padding=0$，对于 $4 \times 4$ 的输入区域，池化核在 $2 \times 2$ 窗口内取最大值，输出尺寸变为 $2 \times 2$。


3. 主要参数
创建 nn.MaxPool2d 实例时，常用参数包括：

kernel_size: 池化核的大小（如 2 或 $(2, 2)$ 表示 $2 \times 2$ 窗口）。
stride: 池化核滑动的步幅（默认等于 kernel_size）。
padding: 输入边缘填充的像素数（默认 0）。
dilation: 池化核元素之间的间距（默认 1，极少使用）。
return_indices: 是否返回最大值的索引（默认 False，用于上采样如 nn.MaxUnpool2d）。
