## 潜空间

### 定义
   没有发音符号的语义空间，每一个维度代表一个独立的语义；如果是图片就是每个通道，通道合在一起才有意义

### 生成

   Word2Vec 使用此方法得到嵌入矩阵
   主要有两种架构：
    
#### 1.CBOW（连续词袋模型）的详细分析   

   ##### 定义
   
   CBOW是一种神经网络模型，旨在通过周围上下文词预测目标词。它属于Word2Vec框架，目标是学习单词的连续向量表示。

   ##### 工作原理
   
   给定一个固定大小的上下文窗口（例如，目标词前后各2个词），CBOW将上下文词的向量求和或平均后，预测中心词。研究显示，其架构包括输入层、隐藏层和输出层：

   ###### 输入层：上下文词用独热编码（one-hot encoding）表示。
   ###### 隐藏层：通过线性变换和非线性激活捕捉语义关系，通常维度为100-300。
   ###### 输出层：生成词汇表上的概率分布，预测目标词。


训练目标：最大化目标词的条件概率，给定其上下文。例如，公式为：
$$J_\theta = \frac{1}{T}\sum_{t=1}^T \log \Pr(w_t | w_{t-n}, \dots, w_{t-1}, w_{t+1}, \dots, w_{t+n})$$
其中，$w_t$是目标词，$n$是窗口半径。
特点：

训练速度：CBOW训练较快，尤其在大型语料库中。研究表明，与Skip-Gram相比，CBOW的收敛时间短，可能只需数小时。
性能：适合频繁词，捕捉句法关系较好，但对稀有词表现较弱，因为其平均化过程可能平滑稀有词的信息。
上下文处理：不考虑上下文词的顺序，符合“词袋”假设（bag of words）。
推荐窗口大小：根据研究，建议窗口大小为5，平衡上下文信息和计算成本。



#### Skip-Gram（跳字模型）的详细分析

##### 定义

Skip-Gram是一种神经网络模型，旨在通过目标词预测其周围上下文词。它也是Word2Vec框架的一部分，强调语义关系的捕捉。

##### 工作原理

给定目标词，Skip-Gram预测窗口内的所有上下文词。例如，目标词为“猫”，可能预测“在”和“屋子里”。其架构与CBOW类似，但输入和输出相反：

###### 输入层：目标词用独热编码表示。
###### 隐藏层：学习目标词的嵌入向量。
###### 输出层：生成上下文词的概率分布。


训练目标：最大化上下文词的条件概率，给定目标词。例如，公式为：

$$J_\theta = \frac{1}{T}\sum_{t=1}^T \sum_{-n \leq j \leq n, j \neq 0} \log \Pr(w_{t+j} | w_t)$$
其中，$n$是窗口半径，损失函数为：
$$-\sum_{i} \sum_{j \in i+N} \log \frac{e^{v_{w_j}' \cdot v_{w_i}}}{\sum_{w'} e^{v_{w'}' \cdot v_{w_i}}}$$

特点：

训练速度：Skip-Gram训练较慢，可能需要数天，尤其在小型语料库中。但在大型语料库中，其计算成本可通过负采样（negative sampling）优化。
性能：对稀有词和短语表现更好，捕捉语义关系更强。例如，研究显示，Skip-Gram在语义相似性任务中优于CBOW。
上下文处理：更重视附近上下文词，权重随距离递减，适合捕捉长距离依赖。
推荐窗口大小：研究建议窗口大小为10，捕捉更广泛的上下文。
   
## 注意力机制 Attention

$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{Q K^T}{\sqrt{d_k}}\right) V$
 
其中 $ \sqrt{d_k} $ 防止点积值过大，稳定梯度。

如何计算Q，K，V

### 步骤：

   1.相似度计算：
   计算 Query 和每个 Key 的点积，得到相关性分数：
   
   score(Q,K i​)=Q⋅K i​
   

   2.缩放，计算分数
   
   $S = \frac{Q K^T}{\sqrt{d^k}}$$
   
   
   3.Softmax 归一化
   把分数转成概率分布：

   𝛼=softmax(score)
   
   
   4.加权求和
   权重乘以对应的 Value 并求和：
                    
   Attention(𝑄,𝐾,𝑉)=𝛼𝑉
   
### 公式整体写法
 
   $\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{Q K^T}{\sqrt{d_k}}\right) V$
