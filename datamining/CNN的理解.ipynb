## CNN的理解

### 举个例子
    计算机图像识别时候，关心的是像素点之间的关系
    
### 如何理解
    模式即特征，有几种款式，并可理解为卷积核，一块可滑动的区域。
    模式的值即为特征值，描述特征的强弱
    
    时域的卷积等于频域的乘积
    
#### 概念 
1. 时域（Time Domain）
    定义：时域描述信号或数据随时间变化的特性。信号在时域中的表示通常是时间（横轴）与信号幅度（纵轴）的关系。
    特点：
    直接展示信号在不同时间点的值。
    反映信号的时间特性，如周期性、突变、趋势等。
    常见于时间序列数据（如音频、心电图、传感器数据）。
    示例：一段音频信号的波形图，横轴是时间，纵轴是振幅。
2. 值域（Value Domain）
    定义：值域通常指信号的幅度、强度或其他特征的分布范围，与时间无关。更广义地，值域可以指信号在某个特定域（如频率域、统计域）中的表示，但在你的问题中，值域更可能是指信号幅度的范围或特性。
    特点：
    关注信号的数值分布或统计特性（如最大值、最小值、均值、方差）。
    不直接依赖时间，而是分析信号的“内容”或“强度”。
    示例：音频信号的幅度直方图，显示振幅值的分布。
    
#### 傅立叶变换

   \begin{equation}
X(f) = \int_{-\infty}^{\infty} x(t) e^{-j2\pi ft} \, dt
\end{equation}


#### 再理解
   卷积定理：时域中的卷积等价于频域中的点乘。
   
   
#### Gabor变换（Gabor Transform）

\begin{equation} G_x(t, f) = \int_{-\infty}^{\infty} x(\tau) g(\tau - t) e^{-j2\pi f \tau} , d\tau \end{equation} 

x(τ)：输入信号（时域）。可理解为原图
\begin{equation} G_x(t, f)  \end{equation} ：Gabor变换的时频表示（时间 t，频率 𝑓）。t可理解为模式的编号，f理解为位置。
𝑔(𝜏−𝑡)：高斯窗函数，中心位于时间 𝑡


#### 再理解

###### 原理：
    Gabor 变换本质上是傅里叶变换的加窗版本，窗函数为高斯函数。
    在频域中，Gabor 滤波器可以看作是对特定频率和方向的频率分量进行滤波。
###### CNN 作用：
    CNN 的卷积核可以模拟 Gabor 滤波器的频域行为，学习特定频率和方向的滤波器。
    例如，CNN 可以学习类似 Gabor 滤波器的卷积核，提取图像的低频（平滑区域）或高频（边缘）特征。

#### 池化层的主要作用

 ######  1.降维（减少计算量）：
    池化层通过下采样减少特征图的空间尺寸（如从 28×28 降到 14×14）。
    降低后续层的参数量和计算复杂度，加速训练和推理。
 ######  2.保留主要特征：
    池化保留特征图中的重要信息（如边缘、纹理），丢弃冗余细节。
    例如，最大池化（Max Pooling）保留局部区域中最强的特征响应。
 ######  3.增强平移不变性：
    池化使 CNN 对输入的微小平移、旋转或缩放更鲁棒。
    例如，图像中的物体移动几个像素，池化后的特征图变化很小。
 ######  4.控制过拟合：
    通过减少特征维度，池化降低模型复杂度，减少过拟合风险。
    类似于正则化效果，帮助模型泛化到新数据。
 ######  5.扩大感受野：
    池化层通过降采样使后续卷积层的卷积核覆盖更大的输入区域，间接扩大感受野。
    例如，两次 2×2 池化后，卷积核能捕捉到原始输入的四倍区域。
