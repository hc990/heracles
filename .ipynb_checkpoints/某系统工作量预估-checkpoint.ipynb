# 背景


# 定义

    规则种类:版本规则、录入规则和签发规则
    三级结构：签发方-系列号-证件种类
    关联规则: 包含并不仅限于：是否电子证件、阅读机版本号、种类代码、证号结构、特殊号码段、
    发行年份、有效期、项目内容、签发地等。
    角色：系统用户、系统管理员、系统高级管理员

# 数据采集模块

## 前台采集数据到后台系统

### 前端系统部分
    
    前端系统数据的预处理，包括采集，打包（功能点数：7？）
    前端系统图片的预处理，包括采集，打包，切割（功能点数：7？）
    前端系统对网络负载的检测（功能点数：3？）
    前端系统对数据的路由选择（功能点数：7？）
    前端系统对数据的下发以及流转（功能点数：7？）
    前端系统对图片的下发以及流转（功能点数：7？）
    前台系统记录采集日志（功能点数：7？）
    
### 后台系统部分

    后台系统采集到每一台终端用户的录入、签发数据的预处理入库（功能点数：7？）
    后台系统采集到每一台终端用户的图片的还原，打标，存储（功能点数：7？）
    后台系统记录采集日志，把数据处理过程，包括并不仅限于“开始时间、验证台号、前台人员账号ID和姓名、采集证件国家和版本，是否完整采集了多种光源下的资料页图片，是否采集到所需的xml文件，是否完整采集到了xml文件中的必要项内容（如有缺失需注明），结束时间”记录与数据库中；（功能点数：7？）
     

## 后台系统反馈采集数据的结果
    
### 前端系统部分

    前端系统给予弹窗（功能点数：7？）
    前端系统记录反馈日志（功能点数：7？）
    
### 后台系统部分    
    后台系统反馈结果（功能点数：7？）
  
## 后台系统可以实时查询保存的信息
    
    用户可以通过关键字查询通过后台系统可以查询到刚刚处理过的数据，以及采集状态，反馈状态等信息（功能点数：7？）   
    
# 规则管理模块

## 版本规则
    
### 关联规则管理
    
    系统管理员可以按照三级结构管理关联规则，样本图片。
    本子模块包含以下功能点：
    系统管理员可以通过三级结构可以搜索到与三级结构匹配的关联规则，输入系列号、证件号有提示功能；（功能点数：5？） 
    系统管理员可以上传并关联样本图片；并可以指定关联规则生效或废止；（功能点数：7？）
    后台系统自动生成版本记录，系列号，默认初始系列号为*010-0。（功能点数：7？）
    系统管理员可以按照关联规则属性模版生成关联规则，并生成相应的版本记录；（功能点数：5？）
    后台系统依据逻辑审核录入规则；（功能点数：7？）
    系统管理员可以按照关联规则属性模版修改更新关联规则；（功能点数：5？）
    系统高级管理员可以生效或废止关联规则；（功能点数3？）
    废止录入规则系统后续批处理操作；（功能点数10？）
    系统后台更新版本记录。（功能点数3？）

### 关联规则属性管理（关联规则模版管理）

    系统管理员可以维护关联规则属性。
    本子模块包含以下功能点：
    系统管理员可以通过三级结构搜索到与三级结构匹配的关联规则模版；（功能点数5？）
    系统管理员可以选定关联规则模版并增加、修改、删除模版属性；（功能点数10？）
    系统高级管理员可以生效或废止关联规则模版 。（功能点数3？）
    系统后台更新版本记录。（功能点数3？）
    
    
### 前台应用规则？

    需求需要明确
